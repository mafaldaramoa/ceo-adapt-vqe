```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

## 32 qubits with XXZ

```python
f = h5py.File("../data/xxz_32_qubits.hdf5", "r")
dmrg_energy = f["dmrg_energy"][()]
adapt_energies = f["adapt_energies"][:]
adapt_times = f["adapt_times"][:]
f.close()
```

```python
adapt_err = np.abs(adapt_energies - dmrg_energy)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(range(1, len(adapt_energies) + 1), adapt_err, '.', label="ADAPT")
# ax[0].hlines(dmrg_energy, 1, len(adapt_energies), color='k', label="DMRG")
ax[0].legend()
ax[0].set_ylabel("Abolsute energy error")
ax[0].set_title("32 qubits")
ax[1].plot(range(1, len(adapt_times) + 1), adapt_times)
ax[1].set_ylabel("Time (ns)")
ax[1].set_xlabel("Iteration")
plt.savefig("xxz_32_qubits_convergence.pdf")
```
